```python
import numpy as np
from scipy import stats

# Extract the column
speeds = combined_df['median_speed_sg'].dropna()

# --- Median + IQR ---
median = np.median(speeds)
q1 = np.percentile(speeds, 25)
q3 = np.percentile(speeds, 75)
iqr = q3 - q1

print("=== Median + IQR ===")
print(f"Median: {median:.4f} µm/s")
print(f"Q1: {q1:.4f}, Q3: {q3:.4f}")
print(f"IQR: {iqr:.4f} µm/s")
print(f"Report as: {median:.3f} [{q1:.3f}–{q3:.3f}] µm/s")
```
